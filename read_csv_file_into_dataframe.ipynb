```
from pyspark.sql.types import *

schema = StructType().add(field='Orderid',data_type=IntegerType())\
                    .add(field='productId',data_type=IntegerType())\
                    .add(field='productname',data_type=StringType())\
                    .add(field='customerid',data_type=IntegerType())\
                    .add(field='Order Date',data_type=DateType())\
                    .add(field='customername',data_type=StringType())\
                    .add(field='customeraddress',data_type=StringType())\
                    .add(field='quantity',data_type=StringType())\
                    .add(field='price',data_type=IntegerType())\
                    .add(field='Sales',data_type=StringType())\
                    .add(field='shipping address',data_type=StringType())\
                    .add(field='billing address',data_type=StringType())\
                    .add(field='inventory',data_type=IntegerType())\
                    .add(field='order status',data_type=StringType())\
                    .add(field='product type',data_type=StringType())\
                    .add(field='product category',data_type=StringType())

df = spark.read.csv(path='dbfs:/FileStore/AllFiles',schema = schema,header=True)
display(df)
df.printSchema()
```
